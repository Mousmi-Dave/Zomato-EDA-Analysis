```python
df_country=pd.read_excel('C:/Users/modave/Downloads/Zomatodataset/Country-Code.xlsx')
df.head()
```

```python
df_country.columns
```

```python
#Merge 2 sheets/tables based on 1 common attribute, how works like joins which table you have focus right join,left join etc
final_df=pd.merge(df,df_country,on ='Country Code',how ='left')
final_df.head(2)
```

```python
#To check datatypes
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
#using values we can only show th counts & not country name
value=final_df.Country.value_counts().values
print(value)
```

```python
# SHow Top 3 country
plt.pie(value[:3],labels=country_names[:3], autopct="%1.2f%%")
```

Observation :  Zomato maximum records & transcation are from India, USA & UK
